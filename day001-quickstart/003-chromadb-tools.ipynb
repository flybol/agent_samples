```python
# 列出向量数据库的collections和记录
import chromadb

DB_FILE_PATH = "../data/chroma_db"


def get_collections(db_file_path: str):
    client = chromadb.PersistentClient(path=db_file_path)
    collections = client.list_collections()
    print(len(collections), "collections:", collections[0].__dict__)


get_collections(DB_FILE_PATH)
```

```python
# 换千问模型，对中文支持好一点
from langchain_chroma import Chroma
from langchain_ollama import OllamaEmbeddings

persist_dir = "../data/chroma_db"
vector_stores = []

embedding = OllamaEmbeddings(model="qwen3-embedding:4b")

score_measures = ["default", "cosine", "12", "ip"]
for score_measure in score_measures:
    collection_metadata = {"hnsw:space": score_measure}
    if score_measure == "default":
        collection_metadata = None

    collection_name = f"my_collection_{score_measure}"
    vector_db = Chroma(
        persist_directory=persist_dir,
        collection_name=collection_name,
        embedding_function=embedding,
        collection_metadata=collection_metadata,
    )
```
